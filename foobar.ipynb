```python
# Level 3: Doomsday Fuel
import math
import fractions
from functools import cmp_to_key

def lcm(a, b):
    return abs(a*b) // abs(math.gcd(a, b))

def rref(mat):
    m = mat
    n = len(m)
    for _ in range(100):
        m = sorted(m, key=cmp_to_key(comp))
        for i in range(n):
            done = False
            for j in range(i):
                if m[i][j] != 0:
                    done = True
                    mult1 = lcm(abs(m[i][j]), abs(m[j][j])) // abs(m[i][j])
                    mult2 = lcm(abs(m[i][j]), abs(m[j][j])) // abs(m[j][j])
                    subtract = (m[i][j] < 0) ^ (m[j][j] > 0)
                    for k in range(n+1):
                        if subtract:
                            m[i][k] = m[i][k] * mult1 - m[j][k] * mult2
                        else:
                            m[i][k] = m[i][k] * mult1 + m[j][k] * mult2
            if done:
                break
    m = sorted(m, key=cmp_to_key(comp))
    for i in range(n, -1, -1):
        for j in range(n-1, i, -1):
            if m[i][j] != 0:
                mult1 = lcm(abs(m[i][j]), abs(m[j][j])) // abs(m[i][j])
                mult2 = lcm(abs(m[i][j]), abs(m[j][j])) // abs(m[j][j])
                subtract = (m[i][j] < 0) ^ (m[j][j] > 0)
                for k in range(n+1):
                    if subtract:
                        m[i][k] = m[i][k] * mult1 -  m[j][k] * mult2
                    else:
                        m[i][k] = m[i][k] * mult1 +  m[j][k] * mult2
    return m

def zeros(x, y):
    return [[0 for i in range(y)] for j in range(x)]
    
def comp(rowl, rowr):
    for i in range(len(rowl)):
        if rowl[i] == 0 and rowr[i] == 0:
            continue
        elif rowl[i] == 0:
            return 1
        elif rowr[i] == 0:
            return -1
    return 0

def solution(m):
    n = len(m)
    if sum(m[0]) == 0:
        return [1] + [0 for i in range(n-1)] + [1]

    res = []
    for x in range(n):
        if sum(m[x]) == 0:
            mat = zeros(n, n+1)
            for i in range(n):
                for j in range(n+1):
                    if j == n:
                        mat[i][j] = -m[i][x]
                    elif j == i:
                        mat[i][j] = m[i][j]-sum(m[i]) if sum(m[i]) > 0 else -1
                    elif j == x:
                        mat[i][j] = 0
                    else:
                        mat[i][j] = m[i][j]
            mat = rref(mat)
            res.append(fractions.Fraction(0) if mat[0][-1] == 0 else fractions.Fraction(abs(mat[0][-1]), abs(mat[0][0])))
            
    base = 1
    for frac in res:
        base = lcm(base, frac.denominator)

    
    for i, frac in enumerate(res):
        res[i] = frac.numerator * base // frac.denominator
    
    res.append(base)
    return res
            

```
